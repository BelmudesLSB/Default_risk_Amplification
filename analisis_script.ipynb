### Lets get all the countrys that GREECE buys from:

```python
df_greece_only = df.loc[:, df.columns.get_level_values(1) == 'GRC']
df_greece_only.head()
# Sum by sectors for each country:
Exports_to_greece_by_sector = df_greece_only.groupby(level=1, axis=0).sum()
#Exports_to_greece = df_greece_only.groupby(level=1, axis=1).sum().sum(axis=1)
Total_Exports_to_greece = Exports_to_greece_by_sector.groupby(level=1, axis=1).sum()
#Exports_to_greece.head()
Total_Exports_to_greece.head()
```

```python
df.head()
```

```python
aux = (df.groupby(level=1, axis=0).sum()).groupby(level=1, axis=1).sum()

num_rows, num_cols = aux.shape

# Loop through the diagonal elements and set them to zero
for i in range(min(num_rows, num_cols)):
    aux.iat[i, i] = 0

Total_Exports = aux.sum(axis=1)
Total_Exports.head()
```

```python
Total_Production = ((df.groupby(level=1, axis=0).sum()).groupby(level=1, axis=1).sum()).sum(axis=1)
```

```python
# merge Total_Exports and Total_Production

merged_df = pd.concat([Total_Exports_to_greece, Total_Production, Total_Exports], axis=1)
merged_df.columns = ['Total_Exports_to_greece', 'Total_Production', 'Total_Exports']
# Calculate the share of exports to Greece in total production
merged_df['Share_Production_to_GRC'] = merged_df['Total_Exports_to_greece'] / merged_df['Total_Production']
merged_df['Share_Exports_to_GRC'] = merged_df['Total_Exports_to_greece'] / merged_df['Total_Exports']
merged_df.head()
```

```python
df_sorted = merged_df.sort_values(by='Share_Exports_to_GRC', ascending=False)
df_sorted.head(20)
```
